```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
x = np.linspace(0,6,1001)

p_normal = stats.norm.pdf(x,3)
c_normal = stats.norm.cdf(x,3)

p_uniform = stats.uniform.pdf(x,3)
c_uniform = stats.uniform.cdf(x,3)

p_t =stats.t.pdf(x,3)
c_t = stats.t.cdf(x,3)

p_f =stats.f.pdf(x,3,3)
c_f = stats.f.cdf(x,3,3)

p_exp =stats.expon.pdf(x,3)
c_exp = stats.expon.cdf(x,3)


fig, ax = plt.subplots(5,2 ,figsize=(8,15))

#normal dis.
ax[0][0].plot(x,p_normal/sum(p_normal),color='r')
ax[0][0].set_title('Normal distribution (PDF)')
ax[0][0].set_ylabel('probability')
ax[0][1].plot(x,c_normal,color='r')
ax[0][1].set_title('Normal distribution (CDF)')

# uniform dis.

ax[1][0].plot(x,p_uniform/sum(p_uniform),'g')
ax[1][0].set_title('uniform distribution (PDF)')
ax[1][0].set_ylabel('probability')
ax[1][1].plot(x,c_uniform,'g')
ax[1][1].set_title('uniform distribution (CDF)')

# t-distribution
ax[2][0].plot(x,p_t/sum(p_t),'y')
ax[2][0].set_title('t-distribution (PDF)')
ax[2][0].set_ylabel('probability')
ax[2][1].plot(x,c_t,'y')
ax[2][1].set_title('t-distribution (CDF)')

ax[3][0].plot(x,p_f/sum(p_f))
ax[3][0].set_title('f-distribution (PDF)')
ax[3][0].set_ylabel('probability')

# f-distribution
ax[3][1].plot(x,c_f)
ax[3][1].set_title('f-distribution (CDF)')

#exponential distribution
ax[4][0].plot(x,p_exp/sum(p_exp),'m')
ax[4][0].set_title('exponential distribution (PDF)')
ax[4][0].set_ylabel('probability')

ax[4][1].plot(x,c_exp,'m')
ax[4][1].set_title('exponential distribution (CDF)')

plt.subplots_adjust(hspace=0.3)
plt.show()
```
